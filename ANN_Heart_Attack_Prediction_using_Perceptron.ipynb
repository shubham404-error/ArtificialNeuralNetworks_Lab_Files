```python
from sklearn.linear_model import Perceptron
import pandas as pd
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv(r'https://raw.githubusercontent.com/shubham404-error/Diabetes_ML_Class-Assignment/main/heart.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
X=data.drop(columns='output',axis=1)
y=data.output
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,stratify=y,random_state=100)
print(X.shape,X_train.shape,X_test.shape)
```

```python
clf = Perceptron(max_iter = 1000, tol=1e-3,eta0 =0.2, random_state=42)
clf.fit(X_train,y_train)
```

```python
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print("The accuracy of the model on validation set is",accuracy_score(y_pred,y_test))
```

```python
from sklearn import metrics
print(metrics.classification_report(y_test,y_pred))
```
